```python
##tish part is used to generate label of train and test set of life
import numpy as np
import os

#path_life = 'C:/Users/Administrator/Desktop/lstm_data/life'
path_life = 'C:/Users/Administrator/Desktop/lstm_work/lstm_data/life_cut'
#path_life = 'D:/Dataset/life-test'
sam_list_unsort = os.listdir(path_life)
#sam_list = sorted(sam_list_unsort, key=lambda x: int(x[6:]))
sam_list = sorted(sam_list_unsort, key=lambda x: int(x[6:-4]))
#print(sam_list)
life_data = []

for sample in sam_list:
    print(sample)
    file_path = os.path.join(path_life,sample)
    f = open(file_path,encoding = "utf-8")
    content = f.readlines()
    f.close()
    #sample_life_data = []
    count = 0
    for i in range(10,40,5):
        sample_life_data = []
        j = i + 5 - 1
        tmp = content[j]
        tmp_split = tmp.split()
        tmp_split_float = [float(x) for x in tmp_split]
        #tmp_split_float_nor = [tmp_split_float[0]/40,tmp_split_float[1]]
        #if tmp_split_float[1] > 1e5:
            #count+=1
        sample_life_data.append(tmp_split_float[1])
        life_data.append(sample_life_data)
    #if count != 0:
        #print(sample)
    #life_data.append(sample_life_data)
    #print(sample, 'has been read')    
print(life_data)
```
