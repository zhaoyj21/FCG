```python
import numpy as np
import cv2
import os
#c_dir_path = 'C:/Users/Administrator/Desktop/Dataset1'
#c_dir_path = 'C:/Users/Administrator/Desktop/lstm_data/split'

c_dir_path = 'D:/newdesktop/Desktop/lstm_work/lstm_data/split_cut'
save_path = 'C:/Users/dell/Desktop/Dataset_cut'
#c_dir_path =  'D:/Dataset/split1/'
#save_path = 'C:/Users/Administrator/Desktop/Dataset1_cut'
#save_path = 'C:/Users/dell/Desktop/Dataset_conti_cut'
os.mkdir(save_path)
p_c_dir_list = os.listdir(c_dir_path)
for p_c_dir in p_c_dir_list:
    dir_path = os.path.join(c_dir_path, p_c_dir)
    dir2_path = os.path.join(save_path, p_c_dir)
    os.mkdir(dir2_path)
    filelist = os.listdir(dir_path)
    for cur_file in filelist:
        img = cv2.imread(os.path.join(dir_path, cur_file))
        file_save_path = os.path.join(dir2_path, cur_file)
        cutimg = img[200:650,:]
        cv2.imwrite(file_save_path,cutimg)
    print(p_c_dir, 'has been cut')
```
