```python
import numpy as np
import cv2
import os
c_dir_path = 'C:/Users/dell/Desktop/Dataset_cut_resize'
save_path = 'C:/Users/dell/Desktop/Dataset_sample_repeat7'
#c_dir_path = 'C:/Users/Administrator/Desktop/Dataset_noslice_cut_resize'
#save_path = 'C:/Users/Administrator/Desktop/Dataset_noslice_sample_repeat'
os.mkdir(save_path)
p_c_dir_list = os.listdir(c_dir_path)
#sample_list = ['9','19','29','39','49','59','69','79']
sample_list = ['10','15','20','25','30','35','40']
#sample_list = ['3','5','7','10','12','15','17','20','22','25','27','30','32','35','37','40']
for p_c_dir in p_c_dir_list:
    dir_path = os.path.join(c_dir_path, p_c_dir)
    dir2_path = os.path.join(save_path, p_c_dir)
    os.mkdir(dir2_path)
    filelist = os.listdir(dir_path)
    for cur_file in filelist:
        if cur_file.split('.')[0] not in sample_list:
            continue
        img = cv2.imread(os.path.join(dir_path, cur_file))
        file_save_path = os.path.join(dir2_path, cur_file)
        cv2.imwrite(file_save_path,img)
    for i in range(41,45):
        cur_file2 = str(i)+'.png'
        file_save_path2 = os.path.join(dir2_path, cur_file2)
        cur_file = '40.png'
        img = cv2.imread(os.path.join(dir_path, cur_file))
        cv2.imwrite(file_save_path2,img)
    print(p_c_dir, 'has been sampled')
```
