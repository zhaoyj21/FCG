```python
import numpy as np
import cv2
import os
c_dir_path = 'C:/Users/dell/Desktop/Dataset_cut'
#c_dir_path = 'C:/Users/Administrator/Desktop/Dataset_conti_cut'
save_path = 'C:/Users/dell/Desktop/Dataset_cut_resize'
#save_path = 'C:/Users/Administrator/Desktop/Dataset_conti_cut_resize'
os.mkdir(save_path)
p_c_dir_list = os.listdir(c_dir_path)
for p_c_dir in p_c_dir_list:
    dir_path = os.path.join(c_dir_path, p_c_dir)
    dir2_path = os.path.join(save_path, p_c_dir)
    os.mkdir(dir2_path)
    filelist = os.listdir(dir_path)
    for cur_file in filelist:
        img = cv2.imread(os.path.join(dir_path, cur_file))
        file_save_path = os.path.join(dir2_path, cur_file)
        pic = cv2.resize(img, (136, 92), interpolation=cv2.INTER_AREA)
        #pic = cv2.resize(img, (100, 56), interpolation=cv2.INTER_AREA)
        cv2.imwrite(file_save_path,pic)
    print(p_c_dir, 'has been resized')
```
